```python
for index, row in exit_notes.iterrows():
    totaldays= pd.Timestamp.today()-pd.Timestamp.today()
    if (pd.notnull(row.exit1)and pd.notnull(row.reopened1)):
        totaldays += (row.reopened1-row.exit1)
    if (pd.notnull(row.exit2)and pd.notnull(row.reopened2)):
        totaldays += (row.reopened2-row.exit2)
    if (pd.notnull(row.exit3)and pd.notnull(row.reopened3)):
        totaldays += (row.reopened3-row.exit3)
    if (pd.notnull(row.exit4)and pd.notnull(row.reopened4)):
        totaldays += (row.reopened4-row.exit4)
    if (pd.notnull(row.exit5)and pd.notnull(row.reopened5)):
        totaldays += (row.reopened5-row.exit5)
    exit_notes.loc[index,'totaldays']=totaldays
```
